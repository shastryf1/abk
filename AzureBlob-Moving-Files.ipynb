```
# SAMPLE Code Format

from azure.storage.blob import BlobServiceClient

# Azure
# Get this from Settings/Access keys in your Storage account on Azure portal
account_name = "YOUR_AZURE_ACCOUNT_NAME"
connection_string = "YOUR_AZURE_CONNECTION_STRING"

# Source
source_container_name = "sourcecontainer"
source_file_path = "soure.jpg"
blob_service_client = BlobServiceClient.from_connection_string(connection_string)
source_blob = (f"https://{account_name}.blob.core.windows.net/{source_container_name}/{source_file_path}")

# Target
target_container_name = "targetcontainer"
target_file_path = "target.jpg"
copied_blob = blob_service_client.get_blob_client(target_container_name, target_file_path)
copied_blob.start_copy_from_url(source_blob)

# If you would like to delete the source file
remove_blob = blob_service_client.get_blob_client(source_container_name, source_file_path)
remove_blob.delete_blob()
```

```
# iterating through a container and copy all files

from azure.storage.blob import BlobServiceClient
from azure.storage.blob import ContainerClient
from datetime import datetime

connection_string = "DefaultEndpointsProtocol=https;AccountName=kumarmovefiles;AccountKey=AgxCk2FQcuG4TVnvZotB0u9HNKiJQsp5w/IaAzSCoM5COkNgyOQPovVOOZshdBIhilQ7QFeCefjiQzKmEqV1hg==;EndpointSuffix=core.windows.net"
container_name = "processed"

target_container = "processed2"

# define blob service client, used for copying files
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

# get current date and time
currentTimeDate = datetime.now()

# get only date in yyyy-mm-dd format
currentDate = currentTimeDate.strftime('%Y%m%d')
print("Currentdate" + currentDate)

# connect using container client
container = ContainerClient.from_connection_string(conn_str=connection_string, container_name=container_name)

print("start")
blob_list = container.list_blobs()


#loop through files in a container, copy all the files
for blob in blob_list:
  
  print(blob.name + '\n')
  blob_name = blob.name  
    
  source_url = f"https://kumarmovefiles.blob.core.windows.net/{container_name}/{blob_name}" 
  print ("source file: " +  source_url)
 
  #append date to file name source_file + YYYMMDD
  new_blob_name =  (blob_name.rpartition('.txt')[0]) + "-" + currentDate  + ".txt"
    
  #target_url = f"https://kumarmovefiles.blob.core.windows.net/{target_container}/{new_blob_name}" 
  
  print ("new file name: " +  new_blob_name)  
    
  print("copying files ..")  
  copied_blob = blob_service_client.get_blob_client(target_container, new_blob_name)
  copied_blob.start_copy_from_url(source_url)  
    
print("end")

```
